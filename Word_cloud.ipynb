```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

## Selecting the dataset

```python
df = pd.read_csv('android-games.csv')
```

```python
df.head()
```

## Select the text and amount of text to use in word cloud

```python
len(df)
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
text = " ".join(cat.split()[1] for cat in df.category)
```

```python
text
```

```python
word_cloud = WordCloud(collocations=False,background_color='white').generate(text=text)
```

```python
from IPython.core.pylabtools import figsize
plt.figure(figsize=[12,10])
plt.imshow(word_cloud,interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
plt.figure(figsize=[12,10])
plt.imshow(word_cloud,interpolation='gaussian')
plt.axis("off")
plt.show()
```

```python
plt.figure(figsize=[12,10])
plt.imshow(word_cloud,interpolation='quadric')
plt.axis("off")
plt.show()
```
